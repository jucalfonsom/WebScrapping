```python
# Web Scrapping del periodico El Espectador de Colombia
# Autor: Juan Camilo Alfonso Mesquida
# Fecha: 17/10/2020


import requests
import pandas as pd
from bs4 import BeautifulSoup


def get_notice():
    
    url = 'https://www.elespectador.com/'
    contador = 1
    noticias_df = pd.DataFrame(columns=['Id', 'Título', 'Link', 'Autor', 'Cuerpo'])
    
    try:
        pagina_elespectador = requests.get(url)
        
        if pagina_elespectador.status_code == 200:
            s = BeautifulSoup(pagina_elespectador.text, 'lxml')
            noticias = s.find_all('div', attrs = {'class': 'Card-title card-title h5'})
            
            if len(noticias) > 0:
                print('Se obtuvieron {0} noticias'.format(len(noticias)))
                
                for noticia in noticias:
                    
                    try:
                        link_noticia = '{0}{1}'.format(url.replace('.com/', '.com'), noticia.a.get('href'))
                        titulo_noticia = noticia.a.get_text()
                        print('Obteniendo información de la noticia {0}'.format(contador))
                        nota = requests.get(link_noticia)
                        s_nota = BeautifulSoup(nota.text, 'lxml')
                        
                        # Extraer autor
                        autor_noticia = s_nota.find_all('span', attrs = {'class': 'Article-Author'})
                        
                        # Extraer cuerpo
                        cuerpo_noticia = s_nota.find_all('p', attrs = {'class': 'font--secondary'})                               
                        
                        noticias_df = noticias_df.append({'Id': contador, 'Título': titulo_noticia, 'Link': link_noticia, 'Autor': autor_noticia, 'Cuerpo': cuerpo_noticia}, ignore_index=True)                               
                    except Exception as e:
                        print('Error al tratar de obtener información de la noticia: {0}'.format(e))
                    finally:
                        contador += 1
                
                
                noticias_df.to_csv(r'C:\Noticias El espectador{0}.csv'.format(date.today()), index = False)

            else:
                print('No fue posible obtener ninguna noticia de {0}'.format(url))
        else:
            print('Error en la obtención de la página {0}'.format(url))
        
    except Exception as e:
        print('Error al tratar de obtener información de {1} de tipo: {0}'.format(e, url))
        
if __name__ == "__main__": 
    get_notice()
```
